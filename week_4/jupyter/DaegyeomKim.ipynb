```python
numbers = "011"

# 소수 찾기
from itertools import permutations
import math

# 소수 판별 함수
def is_prime_number(x):
    # 2부터 x의 제곱근까지의 모든 수를 확인하며
    for i in range(2, int(math.sqrt(x)) + 1):
        # x가 해당 수로 나누어 떨어진다면
        if x % i == 0:
            return False 
    return True
def solution(numbers):

    answer = 0
    nubmers_list = [x for x in numbers]
    numbers_combi = []
    for i in range(len(numbers)):
        numbers_combi.extend(permutations(nubmers_list, i+1))

    temp = set()
    for x in numbers_combi:
        temp.add(int(''.join(x)))
    temp.discard(0)
    temp.discard(1)
    
    for p in temp:
        if is_prime_number(p):
            answer += 1
            
    return answer
```

```python
# 수식 최대화
# expression = "100-200*300-500+20"
expression = "50*6-3*2"

from itertools import permutations
from collections import deque
    
def split_expression(expression):
    cnt = 0
    splited = []
    for i, x in enumerate(expression):
        if x == '+' or x == '-' or x == '*':
            splited.append(expression[cnt:i])
            splited.append(x)
            cnt = i+1
    splited.append(expression[cnt:])
    
    return splited

def solution(expression):
    answer = 0
    opers = ['+', '-', '*']
    
    # 연산 기호와 정수 나누기
    splited = split_expression(expression)
    
    # expression에 없는 연산자 제거
    for op in opers:
        if op not in expression:
            opers.remove(op)
            
    # 연산 기호의 경우의 수 
    combi = []
    combi.extend(permutations(opers))
    
    # 연산 우선 순위 대로 연산한 결과 저장
    for comb in combi:
        stacks = [deque(splited), deque()]
        stack_idx1 = 1
        
        for c in comb:
            stack_idx1 = (stack_idx1+1) % 2
            stack_idx2 = (stack_idx1+1) % 2
            
            while len(stacks[stack_idx1]):
                # print(stacks[stack_idx1])
                # print(stacks[stack_idx2])
                
                item = stacks[stack_idx1].popleft()
                if len(stacks[stack_idx2]) and stacks[stack_idx2][-1] == c:
                    c = stacks[stack_idx2].pop()
                    n = stacks[stack_idx2].pop()
                    item = str(eval(str(n)+c+str(item)))
                stacks[stack_idx2].append(item)
                
        # print(stacks)
        result = stacks[len(opers)%2].pop()
        result = abs(int(result))
        answer = max(answer, result)
    
    return answer

solution(expression)
```
