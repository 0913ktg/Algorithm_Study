```python
import heapq

def more_hot_solution(scoville, K):

    cnt = 0
    heapq.heapify(scoville)
    
    # 가장 작은 값의 스코빌 지수가 K 보다 작을 때 스코빌 지수를 섞는다
    while scoville[0] < K:

      # scoville 값을 섞어주고 넣어준다
      try:
            heapq.heappush(scoville, heapq.heappop(scoville) + (heapq.heappop(scoville) * 2))

      # scoville 값계산하다가 pop기능이 heap이 가지고있는 길이보다 길때 발동되면 
      # = 모든 스코빌 값을들 합쳐도 K 모든 스코빌을 K이상으로 만들 수 없으면 retrun -1 
      except IndexError:
            return -1

      # 반복이 한번 돌때마다 cnt + 1 
      cnt += 1
      
    return cnt
```

```python
print(more_hot_solution( [ 9 , 2 , 3 , 4 , 5 , 1 ] , 9 ))
```

```python
def joystick_solution(name):

    # 조이스틱 조작 횟수 
    answer = 0
    
    # 기본 최소 좌우이동 횟수는 길이 - 1
    min_move = len(name) - 1
    
    for i, char in enumerate(name):
    	# 해당 알파벳 변경 최솟값 추가
        answer += min(ord(char) - ord('A'), ord('Z') - ord(char) + 1)
        
        # 해당 알파벳 다음부터 연속된 A 문자열 찾기
        next = i + 1
        while next < len(name) and name[next] == 'A':
            next += 1
            
        # 기존, 연속된 A의 왼쪽시작 방식, 연속된 A의 오른쪽시작 방식 비교 및 갱신
        min_move = min([min_move, 2 *i + len(name) - next, i + 2 * (len(name) -next)])
        
    # 알파벳 변경(상하이동) 횟수에 좌우이동 횟수 추가
    answer += min_move
    return answer
```

```python
name = "JEROEN"
joystick_solution(name)
```
